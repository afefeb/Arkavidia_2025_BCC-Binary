```python
from prophet import Prophet
import pandas as pd
import numpy as np
from itertools import product

# Fungsi MAPE
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Daftar hyperparameter untuk tuning (dengan opsi seasonality on/off)
param_grid = {
    'changepoint_prior_scale': [0.01, 0.05, 0.1, 0.5],
    'seasonality_prior_scale': [0.1, 1.0, 10.0],
    'holidays_prior_scale': [0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative'],
    'weekly_seasonality': [True, False],
    'daily_seasonality': [True, False],
    'yearly_seasonality': [True, False]
}
param_combinations = list(product(*param_grid.values()))

# Data
df = df_dict['Bawang Merah']
provinsi_list = df.columns[1:]  # Semua kolom kecuali 'Date'

# Variabel untuk menyimpan hasil terbaik dan baseline
best_avg_mape = float('inf')
best_params = None
baseline_mape_list = []

# Grid Search
for params in param_combinations:
    mape_list = []  # Simpan MAPE untuk setiap provinsi

    for provinsi in provinsi_list:
        # Ubah format tanggal
        df_prov = df[['Date', provinsi]].rename(columns={'Date': 'ds', provinsi: 'y'})

        # Ambil 2 bulan terakhir sebagai validasi
        valid_size = 60  # Asumsi sekitar 30 hari per bulan
        train = df_prov.iloc[:-valid_size]  # Semua kecuali 2 bulan terakhir
        valid = df_prov.iloc[-valid_size:]  # 2 bulan terakhir

        # Pastikan validasi tidak kosong
        if valid.empty:
            continue

        try:
            # Train model dengan parameter tuning
            model = Prophet(
                changepoint_prior_scale=params[0],
                seasonality_prior_scale=params[1],
                holidays_prior_scale=params[2],
                seasonality_mode=params[3],
                weekly_seasonality=params[4],
                daily_seasonality=params[5],
                yearly_seasonality=params[6]
            )
            model.fit(train)

            # Prediksi
            future = model.make_future_dataframe(periods=valid_size, freq='D')
            forecast = model.predict(future)
            predicted = forecast['yhat'].iloc[-valid_size:].values  

            # Pastikan jumlah prediksi sesuai
            if len(predicted) != len(valid):
                continue

            # Hitung MAPE
            error = mape(valid['y'].values, predicted)
            mape_list.append(error)

        except Exception:
            continue

    # Hitung rata-rata MAPE untuk kombinasi parameter ini
    if mape_list:
        avg_mape = np.mean(mape_list)

        # Simpan parameter terbaik jika rata-rata MAPE lebih kecil
        if avg_mape < best_avg_mape:
            best_avg_mape = avg_mape
            best_params = params

# Baseline: Prophet Default
for provinsi in provinsi_list:
    df_prov = df[['Date', provinsi]].rename(columns={'Date': 'ds', provinsi: 'y'})

    valid_size = 60
    train = df_prov.iloc[:-valid_size]
    valid = df_prov.iloc[-valid_size:]

    if valid.empty:
        continue

    try:
        # Model Prophet Default
        default_model = Prophet()
        default_model.fit(train)

        # Prediksi
        future = default_model.make_future_dataframe(periods=valid_size, freq='D')
        forecast = default_model.predict(future)
        predicted = forecast['yhat'].iloc[-valid_size:].values  

        if len(predicted) != len(valid):
            continue

        # Hitung MAPE
        error = mape(valid['y'].values, predicted)
        baseline_mape_list.append(error)

    except Exception:
        continue

# Hasil perbandingan
if best_params and baseline_mape_list:
    avg_baseline_mape = np.mean(baseline_mape_list)
    
    print(f"✅ **Parameter terbaik:** {best_params}")
    print(f"📊 **Rata-rata MAPE terbaik:** {best_avg_mape:.4f}")
    print(f"🔹 **Rata-rata MAPE Prophet Default:** {avg_baseline_mape:.4f}")
    
    improvement = ((avg_baseline_mape - best_avg_mape) / avg_baseline_mape) * 100
    print(f"📈 **Peningkatan akurasi dibanding default:** {improvement:.2f}%")
else:
    print("⚠️ Tidak ada parameter yang menghasilkan model valid!")
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler

# Set seed untuk reproducibility
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dictionary untuk menyimpan hasil prediksi
all_predictions = []

# Rentang waktu prediksi
predict_start = '2024-10-01'
predict_end = '2024-12-31'
predict_periods = (pd.to_datetime(predict_end) - pd.to_datetime(predict_start)).days + 1

# Parameter LSTM
seq_length = 30
epochs = 50
batch_size = 16
learning_rate = 0.001

import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable

class LSTMPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=128, num_layers=4, output_size=1):
        super(LSTMPredictor, self).__init__()
        self.num_layers = num_layers
        self.hidden_size = hidden_size

        # LSTM dengan lebih banyak layer dan dropout
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=0.3)

        # Fully connected layers untuk memproses output LSTM
        self.fc_1 = nn.Linear(hidden_size, 64)  
        self.fc_2 = nn.Linear(64, output_size)  
        self.relu = nn.ReLU()

    def forward(self, x):
        # Inisialisasi hidden state dan cell state
        h_0 = Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device))
        c_0 = Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device))

        # Forward pass LSTM
        lstm_out, _ = self.lstm(x, (h_0, c_0))

        # Ambil output terakhir dari LSTM
        out = lstm_out[:, -1, :]
        out = self.relu(self.fc_1(out))  # Fully connected layer pertama
        out = self.fc_2(out)  # Fully connected layer kedua (output harga)

        return out


for komoditas, df in df_dict.items():
    print(f"\n=== Memproses {komoditas} ===")
    
    df['Date'] = pd.to_datetime(df['Date'])
    
    for provinsi in df.columns[1:]:
        print(f"  → Prediksi untuk {provinsi}...")
        
        df_prov = df[['Date', provinsi]].rename(columns={'Date': 'ds', provinsi: 'y'})
        
        # Normalisasi data
        scaler = MinMaxScaler()
        df_prov['y'] = scaler.fit_transform(df_prov[['y']])
        
        # Buat data sekuens
        X, y = [], []
        for i in range(len(df_prov) - seq_length):
            X.append(df_prov['y'].iloc[i:i+seq_length].values)
            y.append(df_prov['y'].iloc[i+seq_length])
        
        X, y = np.array(X), np.array(y)
        X = torch.tensor(X, dtype=torch.float32).unsqueeze(-1).to(device)
        y = torch.tensor(y, dtype=torch.float32).unsqueeze(-1).to(device)
        
        # Pisahkan training & testing
        train_size = int(len(X) * 0.8)
        X_train, X_test = X[:train_size], X[train_size:]
        y_train, y_test = y[:train_size], y[train_size:]
        
        # Bangun model LSTM
        model = LSTMPredictor().to(device)
        criterion = nn.MSELoss()
        optimizer = optim.Adam(model.parameters(), lr=learning_rate)
        
        # Train model
        for epoch in range(epochs):
            model.train()
            optimizer.zero_grad()
            output = model(X_train)
            loss = criterion(output, y_train)
            loss.backward()
            optimizer.step()
        
        # Prediksi masa depan
        model.eval()
        last_sequence = X[-1].unsqueeze(0)  # Data terakhir sebagai input
        future_preds = []
        
        for _ in range(predict_periods):
            with torch.no_grad():
                next_pred = model(last_sequence).item()
            future_preds.append(next_pred)
            
            last_sequence = torch.cat([last_sequence[:, 1:, :], torch.tensor([[[next_pred]]], dtype=torch.float32).to(device)], dim=1)
        
        # Konversi kembali ke skala asli
        future_preds = scaler.inverse_transform(np.array(future_preds).reshape(-1, 1)).flatten()
        
        # Simpan hasil prediksi
        for i, pred in enumerate(future_preds):
            date = pd.to_datetime(predict_start) + pd.Timedelta(days=i)
            all_predictions.append({
                'id': f"{komoditas}/{provinsi}/{date.strftime('%Y-%m-%d')}",
                'price': pred
            })

# Simpan hasil prediksi ke CSV
submission_df = pd.DataFrame(all_predictions)
submission_df.to_csv('submission_lstm_morehidlayer.csv', index=False)

print("\nSemua prediksi selesai dan disimpan di submission_lstm_pytorch.csv!")
```
